```python
import pandas as pd
df = pd.read_csv("EUESGMANUFACTURER-LIGHT.csv")
df
```

```python
isins_with_names = df[["ISIN", "companyLongName"]].groupby(["ISIN", "companyLongName"]).size().reset_index()[["ISIN", "companyLongName"]]
isins_with_names
```

#  Meaningful data

## ESGClassification

- rate how good the company is in a specific sector (water, waste, emissions, ...) based on how many questions of that category it answered by yes
- keep in mind, that maybe some questions don't have yes / no answers and that maybe no questions from a category occur for a company

## Specific questions

### 27/28 Financial instrument type

- rates how compliant the company is, one of 4 possibilities

```python
res = {}
for _, row in isins_with_names.iterrows():
    res[row["ISIN"]] = {
        "name": row["companyLongName"]
    }
res
```

```python
for isin in res.keys():
    count = df[df["ISIN"] == isin]["ISIN"].count()
    res[isin]["question_count"] = count
res
```

```python
categories = df["ESGClassification"].unique()
for isin in res.keys():
    res[isin]["categories"] = {}
    for category in categories:
        rows = df[(df["ISIN"] == isin) & (df["ESGClassification"] == category)]
        no = rows[rows["ESGClassSymbol"] == "ESG3-2 : No"]["ESGClassSymbol"].count()
        yes = rows[rows["ESGClassSymbol"] == "ESG3-1 : Yes"]["ESGClassSymbol"].count()
        count = rows["ESGClassification"].count()
        res[isin]["categories"][category] = {"count": count, "no": no, "yes": yes, "unknown": count - (no + yes)}
res
```
