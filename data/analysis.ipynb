# Metrics

## Compliance category

- Type: one of 0, 1, 2, 3 where 3 is the best
- describes how compliant a fund is with ESG standards, should be visualized as a "traffic light" system

## Sustainability

- Type: float from 0.0 to 1.0 where 1.0 is the best
- describes how sustainable a fund is, is computed by whether and how much a fund invests in sustainable stocks (e.g. with low energy usage, low carbon footprint, ...)
- categories: 
       [
       'MiFID/IDD Target Market - Manufacturer angle (from a client perspective)',
       'Minimum (or planned) Sustainable investment and taxonomy assets allocation breakdown',
       'Scoping according to SFDR annex template'
       'Greenhouse gas emissions',
       'Energy efficiency',
       ]

## Environment

- Type: float from 0.0 to 1.0 where 1.0 is the best
- describes how environmentally friendly a fund is, e.g. how much it invests in stocks with low waste, low water pollution, low impact on biodiversity-sensible areas, ...
       [
       'Biodiversity',
       'Water',
       'Waste',
       'Environmental',
       'Fossil fuels',
       ]

## Social

- Type float from 0.0 to 1.0 where 1.0 is the best
- describes how much the fund invests in socially difficult stocks (e.g. some which work with controversial weapons, which have conflicts with UN laws, which have gender / social gaps, ...)
[
       'Social and employee matters',
       'Social',
]

```python
import json
from functools import reduce
sustainability_categories = ['MiFID/IDD Target Market - Manufacturer angle (from a client perspective)',
                             'Minimum (or planned) Sustainable investment and taxonomy assets allocation breakdown',
                             'Scoping according to SFDR annex template',
                             'Greenhouse gas emissions',
                             'Energy efficiency']
environment_categories = [
       'Biodiversity',
       'Water',
       'Waste',
       'Environmental',
       'Fossil fuels',
       ]
social_categories = [
       'Social and employee matters',
       'Social',
]
cres = {}
for (key, values) in res.items():
    # sustainabilities
    sustainabilities = list(map(lambda x: 0.0 if x < 0.0 else x, map(lambda x: 0.5 if x > 1.0 else x, map(lambda x: 1.0 if x == 'ESG3-1 : Yes' else x, map(lambda x: 0.0 if x == 'ESG3-2 : No' else x, reduce(
        lambda x, y: x + y, [values["categories"][category] for category in sustainability_categories if category in values["categories"]], []))))))
    sustainability = np.mean(sustainabilities) if len(
        sustainabilities) > 0 else 0.0
    # adjust for few or many answered questions
    if len(sustainabilities) < 5:
        sustainability *= (((len(sustainabilities) / 6) - 1) * 0.1) + 1

    # environmental
    environmental_friendlies = list(map(lambda x: 0.0 if x < 0.0 else x, map(lambda x: 0.5 if x > 1.0 else x, map(lambda x: 1.0 if x == 'ESG3-1 : Yes' else x, map(lambda x: 0.0 if x == 'ESG3-2 : No' else x, reduce(
        lambda x, y: x + y, [values["categories"][category] for category in environment_categories if category in values["categories"]], []))))))
    environmental_friendly = np.mean(environmental_friendlies) if len(
        environmental_friendlies) > 0 else 0.0
    environmental_friendly *= 2
    environmental_friendly = min(1.0, environmental_friendly)
    # adjust for few or many answered questions
    if len(environmental_friendlies) < 5:
        environmental_friendly *= (((len(environmental_friendlies) / 6) - 1) * 0.1) + 1

    # social
    social_friendlies = list(map(lambda x: 0.0 if x < 0.0 else x, map(lambda x: 0.5 if x > 1.0 else x, map(lambda x: 1.0 if x == 'ESG3-1 : Yes' else x, map(lambda x: 0.0 if x == 'ESG3-2 : No' else x, reduce(
        lambda x, y: x + y, [values["categories"][category] for category in social_categories if category in values["categories"]], []))))))
    social_friendly = np.mean(social_friendlies) if len(
        social_friendlies) > 0 else 0.0
    social_friendly *= 2
    social_friendly = min(1.0, social_friendly)
    # adjust for few or many answered questions
    if len(social_friendlies) < 5:
        social_friendly *= (((len(social_friendlies) / 6) - 1) * 0.1) + 1

    cres[key] = {
        "name": values["name"],
        "sustainability": sustainability,
        "environment": environmental_friendly,
        "social": social_friendly,
    }

print(json.dumps(cres, indent=2))

acc_environment = 0.0
num_environment = 0
max_environment = 0.0
min_environment = 1.0
for value in cres.values():
    num_environment += 1
    acc_environment += value["environment"]
    if value["environment"] < min_environment:
        min_environment = value["environment"]
    if value["environment"] > max_environment:
        max_environment = value["environment"]
print(acc_environment, num_environment)
print(acc_environment / num_environment)
print(max_environment)
print(min_environment)
```
